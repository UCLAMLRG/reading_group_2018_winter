```python
# load data from Dressel & Farid (2018)
tab = pd.read_csv('./BROWARD_CLEAN_SUBSET.csv')
charge_tab = pd.read_csv('./CHARGE_ID.csv')
```

```python
# dictionary of keys
race = {1:"white", 2:"black",3:"hispanic", 4:"Asian", 5:"Native American", 6:"Other"}
gender = { 0:"male", 1:"female"}
charge_degree = {0: "misdemeanor", 1: "felony"}
recid = {0:"no",1: "yes"}
```

```python
# randomly select 3 people for initial test
np.random.seed(seed=20) # set the same seed each time so this is repeatable
arr = np.random.randint(size=5,low=0,high=999)
print(arr)
n = 1
for i in arr:
    charge_str = charge_tab[charge_tab['charge_id'] == tab['charge_id'].iloc[i]]['mturk_charge_name'].iloc[0]
    age_str = str(tab['age'].iloc[i])
    deg = charge_degree[tab['charge_degree (misd/fel)'].iloc[i]]
    #print((gender[tab['sex'].iloc[i]],age_str,charge_str))
    #print("The defendant is a %s aged %s. They have been charged with: %s." +\ )
    print('Question '+str(n))
    print("The defendant is a %s aged %s. They have been charged with: %s. This crime is classified as a %s. They have been convicted of %s prior crimes. They have %s juvenile felony charges and %s juvenile misdemeanor charges on their record.\n" % 
          (gender[tab['sex'].iloc[i]],age_str,charge_str,deg,tab['priors_count'].iloc[i],tab['juv_fel_count'].iloc[i],tab['juv_misd_count'].iloc[i]) )
    print("Do you think this person will commit another crime within 2 years? ")
    print(recid[tab['two_year_recid'].iloc[i]])
    print("This person's race is: "+race[tab['race'].iloc[i]])
    print("")
    n = n+1
```
